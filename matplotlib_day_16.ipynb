```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Linear Graph
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y, marker='o', linestyle='--')
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(y, marker= '^')
plt.plot(x, marker= 'o')
```

```python
# barplot in matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
c = ['r','g','b','y']
plt.bar(x, y, color = c, width = 0.5)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
c = ['r','g','b','y']
plt.barh(x, y, color = c,height=0.5)
plt.show()
```

```python
# Scatter plot
x = [1,2,3,4,5,6]
y = [5,3,6,2,8,5]
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel('Y')
plt.scatter(x,y,color= 'red')
plt.show()
```
